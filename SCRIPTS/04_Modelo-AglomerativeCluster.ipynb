# MODELO AGLOMERATIVE CLUSTER

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
```

# Carga de la información

```python
df= pd.read_csv("https://raw.githubusercontent.com/lvherrerab/Sectorizaci-n-de-contratos-p-blicos-en-Colombia/main/OUTPUT/muestra_al_35_sector.csv")

```

# Preprocesamiento

```python

# Crear un vectorizador TF-IDF
tfidf_vectorizer = TfidfVectorizer()

# Calcular la matriz TF-IDF
tfidf_matrix = tfidf_vectorizer.fit_transform(df['limpieza_contexto'])

# Obtener los términos (palabras) del vocabulario
terms = tfidf_vectorizer.get_feature_names_out()

# Mostrar la matriz TF-IDF y el vocabulario
matriz = tfidf_matrix.toarray()
print(terms)
```

# Modelo

```python
from sklearn.cluster import AgglomerativeClustering

# Crear una instancia del modelo Agglomerative Clustering
agglomerative = AgglomerativeClustering(n_clusters=24)

# Ajustar el modelo a tus datos
agglomerative.fit(matriz)

# Asignar etiquetas de cluster a cada documento
labels = agglomerative.labels_

# Visualizar las etiquetas asignadas a cada documento
print(labels)

```

# Métrica de desempeño

## Inercia

```python
from sklearn.metrics import silhouette_score

#Silhouette Score: mide cuán similar es un objeto a su propio cluster (cohesión) en comparación con otros clusters 
#(separación). Una puntuación de silueta más alta indica una mejor calidad de agrupación

silhouette_avg = silhouette_score(tfidf_matrix, agglomerative.labels_)


print("Silhouette Score del modelo Agglomerative Clustering :", silhouette_avg)
```
